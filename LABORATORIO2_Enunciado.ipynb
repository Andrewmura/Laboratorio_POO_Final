# Tratamiento de datos nulos

##### Para los datos nulos categoricos se planetea hacer el tratamiento de datos nulos por medio de la moda en cada columna y para los númericos poder llenarlo con la media o el promedio de los datos

```python
from statistics import mode
```

```python

moda_examen=mode(dfsiu["ESTU_VECESPRESENTOEXAMEN"])
moda_internet=mode(dfsiu["FAMI_TIENEINTERNET"])
moda_computador=mode(dfsiu["FAMI_TIENECOMPUTADOR"])
moda_libros=mode(dfsiu["FAMI_NUMLIBROS"])
moda_ingresos=mode(dfsiu["FAMI_INGRESOFMILIARMENSUAL"])
moda_bilingue=mode(dfsiu["COLE_BILINGUE"])
moda_estrato=mode(dfsiu["ESTU_NSE_INDIVIDUAL"])
```

```python

dfsiu["ESTU_VECESPRESENTOEXAMEN"]=dfsiu["ESTU_VECESPRESENTOEXAMEN"].fillna(moda_examen)
dfsiu["FAMI_TIENEINTERNET"]=dfsiu["FAMI_TIENEINTERNET"].fillna(moda_internet)
dfsiu["FAMI_TIENECOMPUTADOR"]=dfsiu["FAMI_TIENECOMPUTADOR"].fillna(moda_computador)
dfsiu["FAMI_NUMLIBROS"]=dfsiu["FAMI_NUMLIBROS"].fillna(moda_libros)
dfsiu["FAMI_INGRESOFMILIARMENSUAL"]=dfsiu["FAMI_INGRESOFMILIARMENSUAL"].fillna(moda_ingresos)
dfsiu["COLE_BILINGUE"]=dfsiu["COLE_BILINGUE"].fillna(moda_bilingue)
dfsiu["ESTU_NSE_INDIVIDUAL"]=dfsiu["ESTU_NSE_INDIVIDUAL"].fillna(moda_estrato) 
dfsiu["ESTU_NSE_ESTABLECIMIENTO"]=dfsiu["ESTU_NSE_ESTABLECIMIENTO"].fillna(dfsiu["ESTU_NSE_ESTABLECIMIENTO"].mean())
```

```python
def visualizacion_resultados(history):
  epochs = [i for i in range(5)]
  fig, ax = plt.subplots(1,2)
  train_acc = history.history["accuracy"]
  train_loss = history.history["loss"]
  val_acc = history.history["val_accuracy"]
  val_loss = history.history["val_loss"]
  fig.set_size_inches(16, 9)

  ax[0].plot(epochs, train_acc, "go-", label =" Entrenamiento accuracy")
  ax[0].plot(epochs, val_acc, "ro-", label= "Validacion accuracy")
  ax[0].set_title("Entrenamiento & validación accuracy")
  ax[0].legend()
  ax[0].set_xlabel("Epochs")
  ax[0].set_ylabel("Accuracy")

  ax[1].plot(epochs, train_loss, "go-", label =" Entrenamiento Loss")
  ax[1].plot(epochs, val_loss, "ro-", label= "Validacion Loss")
  ax[1].set_title("Entrenamiento & validación Loss")
  ax[1].legend()
  ax[1].set_xlabel("Epochs")
  ax[1].set_ylabel("Loss")

  plt.show()
```

```python
visualizacion_resultados(history)
```
